```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 3.3.4

```python
norm = np.random.normal(10, 10, 100)
plt.hist(norm)
plt.axvline(norm.mean(), color='black', linewidth=2)
plt.axvline(norm.mean()-norm.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(norm.mean()+norm.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
# mean and std are both useful here
```

```python
bi = np.random.binomial(10, 0.4, 100)
plt.hist(bi)
plt.axvline(bi.mean(), color='black', linewidth=2)
plt.axvline(bi.mean()-bi.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(bi.mean()+bi.std(), color='red', linestyle='dashed', linewidth=2)
plt.axis([0,10,0,40])
plt.show()
# mean and std are both useful here - but less useful as p approaches 0 or 1
# QUESTION: why do gaps appear in most of these histograms?
```

```python
poiss = np.random.poisson(4, 100)
plt.hist(poiss)
plt.axvline(poiss.mean(), color='black', linewidth=2)
plt.axvline(poiss.mean()-poiss.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(poiss.mean()+poiss.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
# mean and std become increasingly useful as lambda increases
```

```python
chi = np.random.chisquare(10,100)
plt.hist(chi)
plt.axvline(chi.mean(), color='black', linewidth=2)
plt.axvline(chi.mean()-chi.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(chi.mean()+chi.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
# mean and std become increasingly useful as degree of freedom increases
```

```python
logistic = np.random.logistic(0,1,100)
plt.hist(logistic)
plt.axvline(logistic.mean(), color='black', linewidth=2)
plt.axvline(logistic.mean()-logistic.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean()+logistic.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
# mean and std are both useful metrics
```

```python
pareto = np.random.pareto(10,100)
plt.hist(pareto)
plt.axvline(pareto.mean(), color='black', linewidth=2)
plt.axvline(pareto.mean()-pareto.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean()+pareto.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
# mean might be a little bit useful here; std certainly isn't
```

<strong>Analysis: </strong> Mean and standard deviation become decreasingly useful the more heavily skewed a distribution is.

```python
norm1 = np.random.normal(5,.5,10000)
norm2 = np.random.normal(10,1,10000)
norm_sum = norm1 + norm2
plt.hist(norm_sum)
plt.axvline(norm_sum.mean(), color='black', linewidth=2)
plt.axvline(norm_sum.mean()-norm_sum.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(norm_sum.mean()+norm_sum.std(), color='red', linestyle='dashed', linewidth=2)
plt.show()
print("mean: " + str(norm_sum.mean()))
print("std: " + str(norm_sum.std()))

# It's very unclear what "Evaluate the descriptive statistics against the data" in the instructions means...
# at any rate, the mean is just the sum of the means of the original two datasets.  The standard deviation
# is the square root of the sum of the squared original standard deviations (i.e., the variances simply add)
```

```python
pop1 = np.random.binomial(10,0.2,10000)
pop2 = np.random.binomial(10,0.5,1000)
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print("sample 1 mean: {}".format(sample1.mean()))
print("sample 2 mean: {}".format(sample2.mean()))
print("sample 1 std: {}".format(sample1.std()))
print("sample 2 std: {}".format(sample2.std()))
```

<strong>1) increasing n to 1000 </strong><br><br>
EXPECTATIONS:<br>
Means will stay almost the same, converging to the true means of the populations, which are very close to 2.0 and 5.0<br>
STDs will decrease slightly?<br><br>
RESULTS:<br>
Means as expected<br>
STDs actually increased slightly.  After multiple tests, there's no clear and consistend difference.

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

print("sample 1 mean: {}".format(sample1.mean()))
print("sample 2 mean: {}".format(sample2.mean()))
print("sample 1 std: {}".format(sample1.std()))
print("sample 2 std: {}".format(sample2.std()))
```

<strong>2) increasing p1 to 0.3, then decreasing p2 to 0.4 </strong><br><br>
EXPECTATIONS:<br>
pop1 mean will increase; pop2 mean will decrease<br>
STD1 will increase; STD2 will decrease slightly with both changes, t-value will decrease because the difference between means is smaller (overruling the increase in standard error, in the case of increasing p1), and p-value will increase since the chance of a difference in sample means occurring in the same population is greater the smaller the difference is<br><br>
RESULTS:<br>
Everything as expected.

```python
pop1 = np.random.binomial(10,0.3,10000)
pop2 = np.random.binomial(10,0.4,1000)
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
diff = sample1.mean()-sample2.mean()
SE = (sample1.std()**2/100 + sample2.std()**2/100)**0.5
print(abs(diff/SE))

print("sample 1 mean: {}".format(sample1.mean()))
print("sample 2 mean: {}".format(sample2.mean()))
print("sample 1 std: {}".format(sample1.std()))
print("sample 2 std: {}".format(sample2.std()))

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

<strong> 3) Changed distribution to Poisson </strong><br><br>
EXPECTATIONS:<br>
Mean won't be as good a measure of center for a lower lambda value, like 3, because it will be
skewed right, but it should still capture the probabilities effectively<br><br>
RESULTS:<br>
I don't have a clean way to evaluate the probabilistic measure, but the prediction about skew and center was correct.

```python
pop1 = np.random.poisson(3,10000)
pop2 = np.random.poisson(6,10000)
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=0.5)
plt.hist(sample2, alpha=0.5)
plt.show()

print("sample 1 mean: {}".format(sample1.mean()))
print("sample 2 mean: {}".format(sample2.mean()))
print("sample 1 std: {}".format(sample1.std()))
print("sample 2 std: {}".format(sample2.std()))

from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```
